```python
anger_list = []
contempt_list = []
fear_list = []
disgust_list = []
happiness_list = []
neutral_list = []
sadness_list = []
surprise_list = []
```

```python
# a dictionary of ists, one list containing each person's emotions for that particular emotion
emot_dict = {
    'anger_list': [],
    'contempt_list' : [],
    'fear_list' : [],
    'disgust_list' : [],
    'happiness_list' : [],
    'neutral_list' : [],
    'sadness_list' : [],
    'surprise_list' : []
    }
        
```

```python
emot_dict['anger_list']
```

```python
for i in scores:
    emot_dict['anger_list'].append(i['anger'])
    emot_dict['contempt_list'].append(i['contempt'])
    emot_dict['fear_list'].append(i['fear'])
    emot_dict['disgust_list'].append(i['disgust'])
    emot_dict['happiness_list'].append(i['happiness'])
    emot_dict['neutral_list'].append(i['neutral'])
    emot_dict['sadness_list'].append(i['sadness'])
    emot_dict['surprise_list'].append(i['surprise'])
```

```python
emot_dict
```

```python
# for example, access the anger_list:
emot_dict['anger_list']
```

```python
# average of anger_list
np.mean(emot_dict['anger_list'])
```

```python
# averag of happiness
np.mean(emot_dict['happiness_list'])
```

```python
# print all emotions with list of emotion's value
for i in emot_dict:
    print(str(i), str(emot_dict[i]), '\n')
```

```python
# print out the average score emotion value for each emotion

for emo_list in emot_dict:
    print(emo_list)
    print('Avg: ' + str(np.mean(emot_dict[emo_list])))
    print('Std: ' + str(np.std(emot_dict[emo_list])))
    print(emot_dict[emo_list])
    print('\n')
```
